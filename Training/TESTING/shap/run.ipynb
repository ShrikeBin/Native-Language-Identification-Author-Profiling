```python
from load_models import load_models

models = load_models()
```

```python
# models are loaded, explainers are loaded
import shap
import matplotlib.pyplot

shap.initjs()
def explain(text):
    for model in models:
        print(model.prediction_string(text))
        shap.plots.force(*model.explain(text), matplotlib=True)
```

```python
explain("saminamina e-e waka waka I'm gay")
```

```python
explain("I love big black cars, i love 'em roar and scream all around my hood. Anyways Dacia sandero is back in sale for the left hand drive marked, did you know that? - i didnt, now i do XD")
```
